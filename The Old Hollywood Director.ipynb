```python
import pandas as pd
import numpy as np
%cd ~/pandas/
```

```python
actors = pd.read_excel('actors_unformatted.xlsx')
movies = pd.read_excel('movies_unformatted.xlsx')
```

```python
actors.head(10)
```

```python
actors = actors.reset_index().fillna(method='ffill')
actors.head()
```

```python
actors = actors.reindex([x[0] for x in actors.groupby('level_0').groups.values()])
actors.head()
```

```python
actors = actors.drop('level_1',1).set_index(['level_0'])
actors.head()
```

```python
actors.index.names = ['imdb_movie_id']
actors.head()
```

```python
movies = movies.set_index(['imdb_movie_id'])
movies.head()
```

```python
actors.stack().reset_index().set_index(0).drop('level_1',1).reset_index().groupby(0).apply(lambda x: list(x['imdb_movie_id']))
```

```python
actors.head()
```

```python
x = actors.stack().reset_index().drop('level_1',1).set_index(0).reset_index().groupby(0).apply(lambda x: sum([movies.ix[y]['revenue'] for y in list(x['imdb_movie_id'])]))
```

```python
x.columns = ['revenue']
```

```python
x.index.names = ['Actor']
```

```python
x.order(ascending=False)[:20]
```
