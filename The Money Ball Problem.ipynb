```python
batsmen = pd.read_csv('ipl_player_hunt_batsmen.csv',na_values='-')
bowlers = pd.read_csv('ipl_player_hunt_bowlers.csv',na_values='-')
```

```python
# I know cricinfo data sucks, but we got to play with the cards we are dealt
batsmen.head()
```

```python
batsmen = batsmen[(batsmen.no_of_matches > 5) & (batsmen.balls_faced > 30)]
batsmen.head()
```

```python
bowlers = bowlers[(bowlers.no_of_matches > 5) & (bowlers.no_of_overs > 10)]
bowlers.head()
```

```python
batsmen['value_as_batsmen'] = np.int64(100 * ((batsmen.batting_strike_rate * batsmen.no_of_runs)/batsmen.no_of_matches) + (0.1 * batsmen['100s']) + (0.2 * batsmen['50s']) - (0.5 * batsmen['ducks']))
bowlers['value_as_bowler'] = np.int64(7500 * ((bowlers.bowling_strike_rate * bowlers.no_of_wkts)/bowlers.no_of_matches) + (0.5 * bowlers.no_of_maidens) + (0.25 * bowlers['four_wkts_in_a_innings']) + (0.16 * bowlers['five_wkts_in_a_innings']) - (0.5 * bowlers['no_of_runs_given']))
```

```python
batsmen = batsmen.set_index(['player_name'])
bowlers = bowlers.set_index(['player_name'])
```

```python
print batsmen.count()
print bowlers.count()
```

```python
players_joined = batsmen.join(bowlers,lsuffix='_batsmen',how='outer')
players_joined.head()
```

```python
players_merged = pd.merge(batsmen.reset_index(),bowlers.reset_index(),on=['player_name','player_country'],how='outer',suffixes=('_batsmen','_bowlers'))
players_merged.head()
```

```python
players_concatenated = pd.concat([batsmen,bowlers],join='outer')
players_concatenated.head()
```
